## PLOTTING IN PYTHON

Here, a simple introductory example of how to plot in Python will be shown.
You should get used to import the modules 'matplotlib.pyplot' and 'numpy', which makes plotting quite easy:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,1,50)
y = x**2

plt.plot(x,y)            # Plot y vs x using the 'plot(x,y)' method.
plt.xlabel('x')          # Label the x-axis whatever you like. Argument must be a string.
plt.ylabel('y')          # Label the y-axis whatever you like. Argument must be a string.
plt.title('$y = x^2$')   # Choose a title for your figure. Must be a string. Use $$ if you want to use
                         #     sub- or super-indices, or greek letters. Same syntax as LaTex.
plt.show()               # The command plt.show() is needed when running from the command line, which is what we are doing in Jupyter Notebooks...
```

When using \\$ inside strings, we are telling matplotlib that we want to use LaTex syntax. This means that 'x^2' will show as exactly that (i.e 'x^2'), while '\\$x^2\\$' will be shown as $x^2$. This only works when using matplotlib, your regular strings in Python (such as paths to files) will not be affected by a \$ symbol.

Below is a slightly more advanced example.
With matplotlib.pyplot, we can actually plot anything we want to, from bar-, pie-, 3D-plots etc.
But most of the time, we only need the following commands:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,1,50)
y_1 = x                  # y_1 is linear in x
y_2 = x**2               # y_2 is quadratic in x
y_3 = x**3               # y_3 is cubic in x

plt.plot(x,y_1, 'r', label='$y=x$')      # Choose red color, and label the line 'y=x'
plt.plot(x,y_2, 'b', label='$y=x^2$')    # Choose blue color, and label the line 'y=x^2'
plt.plot(x,y_3, 'g', label='$y=x^3$')    # Choose green color, and label the line 'y=x^3'
plt.legend(loc='best')                   # Show all the labels for the lines you have plotted

plt.xlabel('x')         # Label the x-axis with a name 
plt.ylabel('y')         # Label the y-axis with a name
plt.xlim([0,1])         # Set the limits of the x-axis 
plt.ylim([0,1])         # Set the limits of the y-axis 
plt.title('Different polynomials')   
plt.show()
```

We will now conclude this very short introduction to plotting in Python by showing how to save our plots.

## READING FROM AND WRITING TO FILES

'open(filename, mode)' is used to interact with files. Both arguments (both 'filename' and 'mode') should be strings.
'filename' should be the path (or simply the name of the file if you are in the same folder already) to the file you want to interact with, and 'mode' will specify how you want to interact with it.

Typical values for 'mode' are 'r' (read from file) and 'w' (write to file). There are other modes, like 'a' (append new lines at the bottom of the file), and 'r+' and 'w+'. You will most likely only need 'r' and 'w'.

When reading/writing files, you need to close the file when you are done. Most of the time, nothing will happen if you forget.
BUT, keeping the file open in memory will use space in RAM, impacting performance. Windows also treats open files as 'locked', meaning you can't access them while they are open somewhere else. Edits done to a file will also only be saved when the file is closed. If you are a gamer, you might remember that you are always told to not turn off the power while saving your game. 
If the power is turned off while files are written to, the file will be deleted. Files in Python behave likewise. You should therefore always close your files (thus saving your edits) when you are done.

The most basic way to interact with files are:

```python
#    file = open(filepath, 'r')                    # Open your file here
#        ... do something with your file here ...
#    file.close()                                  # And close your file here 
```

A better way to interact with files are:

```python
#    with open(filepath, 'r') as file:             # Open your file here 
#        ... do something with your file here ...         
#                                                  # Automatically closed here
```

When opening the file with the 'with' statement, it is automatically closed when there is no more indented code!
This way, you will never forget to close your file. This is the recommended way to interact with files in Python.

Below, we will show two examples of how to interact with files. First, we will write some values to a file.
Then, we will read this file, and verify that the values are indeed the same that we wrote.
We will write a file that looks like this:

```python
#    first value   : 20.123
#    second value  : 34.2
#    last value    : 50.17
```

The file is now written using the following code:

```python
with open('testingWriteRead.txt', 'w') as file:                # A file named 'testing.txt' is created, and written to using mode 'w'
    file.write('first value   : {0}\n'.format(20.123))      # Remember that '\n' is the 'new line' statement...
    file.write('second value  : {0}\n'.format(34.2))
    file.write('last value    : {0}'.format(50.17))
```

If you now take a look in the same folder as you opened this .ipynb file from, you will see a new file, called 'testing.txt'.
If you open it (install Notepad++, it is perhaps the best editor for .txt files on Windows), you can verify that the file does indeed contain the same lines as described above. 

Link to the current version of Notepad++ download below:

https://notepad-plus-plus.org/download/v7.7.1.html

We will now read the 'testing.txt' file, and store the content in the variable 'values':

```python
values = []
with open('testingWriteRead.txt', 'r') as file:
    for line in file:
        line = line.split(':')        # Split the line each position a '=' symbol is encountered
        values.append(float(line[1])) # Store the number in a list named 'values'
                                      # The file is automatically closed here
```

The values from the file 'testing.txt' were now read and stored in the variable 'values'.

## PRINTING TO SCREEN (Useful to know, but not necessarily for this course)

There are several ways to prepare the output that can be printed to screen. The three examples below gives identical output:
    1. Separate different expressions using the comma ',' inside print(), e.g. 
             print("The number is", 13").
    2. Use string formatting, which is combining different expressions into a single string, and then print that string.
       Two different ways of doing that: Either use '%', or use '.format()', both give identical results:
             print("The number is %i" % 13)          # Specify the type of the input. Here, we are using '%i' for 'integer'.
             print("The number is {}".format(13))    # We don't need to specify the input type when using '.format()'

It is recommended to use the new-style '.format()' instead of the old-style '%' when using string formatting.
As a final example, we will show how to print the file that was read and stored in 'values':

```python
print('first value   : {}'.format(values[0]))    
print('second value  : {}'.format(values[1]))    
print('last value    : {}'.format(values[2]))
```

```python
print('first value   :', values[0])    
print('second value  :', values[1]) 
print('last value    :', values[2])
```

As you can see, the output from 'print()' is identical. Choose whichever way to print that you are the most comfortable with.
